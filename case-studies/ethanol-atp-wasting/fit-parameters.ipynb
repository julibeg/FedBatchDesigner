## Case study: Ethanol production in ATP-wasting _S. cerevisiae_

[Zahoor et al. (2020)](https://doi.org/10.1186/s13068-020-01822-9) enforced ATP wasting in an _S. cerevisiae_ strain by introducing genes for heterologous expression of ATPase subunits.
Since ATP generation is tied to ethanol formation in anaerobic conditions, this should lead to increased ethanol production, especially under growth-limiting conditions (like nitrogen starvation).
We can use [FedBatchDesigner](https://chemnettools.anc.univie.ac.at/FedBatchDesigner/) to explore the titer&ndash;productivity trade-off of a two-stage fed-batch process using the ATP-wasting strain with nitrogen starvation in the second stage.
In order to do this, we need to estimate the parameters determining product formation ($\frac{dp}{dt} = x (\pi_0 + \mu \pi_1)$) from the data in the paper.
Note that parameters determined from shake flask data of course have very limited transferability to bioreactor conditions.
However, we can use this as a starting point to inform the first fed-batch experiments in a scale-up study.

We will use the data from Fig. 2 to fit $pi_0$ and $pi_1$ under normal conditions (i.e. without nitrogen starvation).
$pi_0$ under nitrogen-starved conditions can be taken from Table 3.
The remaining parameters can be taken from the literature (ATP yield and biomass yield on glucose in anaerobic conditions) or assumed for a conventional lab-scale bioreactor setup ($V_{batch}$, $V_{end}$, etc).


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

```python
df_orig = pd.read_csv("data-from-fig-2.csv", index_col=0)
# drop the last data points when substrate is running out
df = df_orig.loc[:7].copy()
# rename columns for convenience
df.columns = ["x", "p", "s"]
# we don't need the glucose concentration
df = df[["x", "p"]]
# convert the ethanol mM to g/L
df["p"] = df["p"] / 1000 * 46
```

Let's have a look at the parsed data.

```python
fig, ax = plt.subplots()
ax2 = ax.twinx()

ax.plot(df["x"], "bo--", label="Biomass")
ax2.plot(df["p"], "ro--", label="EtOH")

ax.set_xlabel("Time [h]")
ax.set_ylabel("Biomass [g/L]")
ax2.set_ylabel("EtOH [g/L]")
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines += lines2
labels += labels2
ax.legend(lines, labels, loc="upper left")

for ax in fig.axes:
    ax.set_ylim(0, ax.get_ylim()[1])
```

At first we attempt to fit an exponential growth curve to the biomass concentration.

```python
# fit exponential growth to the data
t = df.index.values
t_interp = np.linspace(df.index[0], df.index[-1], 101)


def fit_x_exp(t, x0, mu):
    return x0 * np.exp(mu * t)


(x0_fit, mu_fit), _ = scipy.optimize.curve_fit(
    fit_x_exp,
    t,
    df["x"],
    p0=[0.3, 0.1],
)

x_fit = fit_x_exp(t_interp, x0_fit, mu_fit)

plt.plot(t, df["x"], "ko--", label="Biomass data")
plt.plot(t_interp, x_fit, "r--", label="Exponential growth fit")
plt.ylabel("Biomass [g/L]")
plt.xlabel("Time [h]")
plt.legend()
plt.show()
```

Due to the lag phase in the first few hours the fit is not great. However, we can instead also just interpolate the biomass concentrations with cubic splines and use those for fitting the ODE for ethanol formation later.

```python
# interpolate growth with cubic splines
x_cs = scipy.interpolate.CubicSpline(df.index, df["x"], bc_type="natural")
# let's also get the first derivative of the spline (we can use this later to calculate
# mu at any point in time)
x_cs_d1 = x_cs.derivative(1)

# let's visualize the interpolation
x_interp = x_cs(t_interp)

plt.plot(df.index, df["x"], "ko--", label="Biomass data")
plt.plot(t_interp, x_interp, "r--", label="Interpolated spline fit")
plt.ylabel("Biomass [g/L]")
plt.xlabel("Time [h]")
plt.legend()
plt.show()
```

Now that we can interpolate $x$ and $\mu$, we use them to integrate the ODE for $p$ in order to fit $\pi_0$ and $\pi_1$ (according to $\frac{dp}{dt} = x (\pi_0 + \mu \pi_1)$)

```python
def predict_p(t, pi_0, pi_1, p0=0):
    # define an ODE for the ethanol concentration
    def ODE(t, p):
        x = x_cs(t)
        mu = x_cs_d1(t) / x
        return x * (pi_0 + mu * pi_1)

    # solve the ODE
    res = scipy.integrate.solve_ivp(
        fun=ODE,
        t_span=(t[0], t[-1]),
        y0=[p0],
        t_eval=t,
    )
    return pd.Series(res.y[0], index=res.t, name="EtOH [g/L]")


def get_residuals(params):
    pi_0, pi_1 = params
    p_pred = predict_p(t, pi_0, pi_1)
    return df.loc[t, "p"] - p_pred


res = scipy.optimize.least_squares(
    get_residuals,
    x0=[0.1, 0.1],
    bounds=([0, 0], [np.inf, np.inf]),
)

pi_0, pi_1 = res.x
print(f"stage 1: {pi_0=:.3g}, {pi_1=:.3g}")
```

Let's plot the predicted ethanol concentrations with these parameters and the interpolated biomass concentrations.

```python
# check the fit of product concentration visually
plt.plot(df.index, df["p"], "ko--", label="EtOH data")
plt.plot(t_interp, predict_p(t_interp, *res.x), "r--", label="EtOH fit")
plt.ylabel("EtOH [g/L]")
plt.xlabel("Time [h]")
plt.legend()
plt.show()
```

This doesn't look too bad! We can also check if this gives us a similar value to the $r_{Eth}= 28.77$ mmol/(g CDW h) reported in Table 2 for the exponential phase between 2.5 and 7 hours (at a $\mu$ of 0.235 /h).
> Note that $r_{Eth}$ is the same as $\pi = \pi_0 + \mu \pi_1$ in the notation of our framework.

```python
r_Eth_table = 28.8 / 1000 * 46
r_Eth_fitted = pi_0 + 0.235 * pi_1
print(f"{r_Eth_table=:.3f}, {r_Eth_fitted=:.3f}")
```

This looks close enough!
